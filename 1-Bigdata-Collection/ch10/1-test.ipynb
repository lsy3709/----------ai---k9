```python
import bs4
webPage = open(r'./Sample02.html', 'rt', encoding='utf-8').read()
bsObject = bs4.BeautifulSoup(webPage, "html.parser")

print(bsObject.find('div'))      # 첫 div 태그
print(bsObject.find('li'))       # 첫 li 태그
for li in bsObject.findAll('li'):
    print(li.text.strip())       # 한빛출판네트워크 / 비기너 / 데이터 분석
```

```python
import bs4
html = '''<HTML><BODY>
  <div id='myId1'> 아기공룡 </div>
  <div class='myClass1'> 내 친구 </div>
  <a href="http://www.daum.net"> 다음 바로가기 </a>
  <a href="http://www.naver.com"> 네이버 바로가기 </a>
</BODY></HTML>'''
bs = bs4.BeautifulSoup(html, "html.parser")

print(bs.find('div', {'id': 'myId1'}).text)     #  아기공룡 
print(bs.find('div', {'class': 'myClass1'}).text)  #  내 친구 

for a in bs.findAll('a'):
    print(a.text.strip(), "→", a['href'])       # 텍스트와 링크 주소 함께
```

```python
import bs4, urllib.request, csv, datetime, time

url = "https://news.nate.com/weather?areaCode=11H20201"   # 서울 날씨
html = urllib.request.urlopen(url).read()
bs = bs4.BeautifulSoup(html, 'html.parser')

tag = bs.find('div', {'class': 'right_today'})
temper = tag.find('p', {'class': 'celsius'}).text
humi   = tag.find('p', {'class': 'humidity'}).text
rain   = tag.find('p', {'class': 'rainfall'}).text
wind   = tag.find('p', {'class': 'wind'}).text
print(temper, humi, rain, wind)

# CSV 저장 (헤더 한 번 + 데이터 추가)
with open('weather.csv', 'w', newline='', encoding='utf-8-sig') as f:
    csv.writer(f).writerow(['연월일','시분초','온도','습도','강수량','풍향'])
now = datetime.datetime.now()
row = [now.strftime('%Y-%m-%d'), now.strftime('%H:%M:%S'), temper, humi, rain, wind]
with open('weather.csv', 'a', newline='', encoding='utf-8-sig') as f:
    csv.writer(f).writerow(row)
print('저장:', row)
```

```python
import bs4, urllib.request, csv, json

url = "https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=TodayHot&BranchType=1&CID=336"
bs = bs4.BeautifulSoup(urllib.request.urlopen(url).read(), 'html.parser')

book_data = []
for i, book in enumerate(bs.findAll('div', {'class': 'ss_book_list'})):
    if i % 2 != 0:
        continue                       # 책 하나당 두 번 나와서 짝수 인덱스만
    ul = book.find('ul')
    title = ul.find('a', {'class': 'bo3'}).text
    lis = ul.findAll('li')
    if len(lis) == 5:
        price, author = lis[3].find('span').text, lis[2].find('a').text
    elif len(lis) == 4:
        price, author = lis[2].find('span').text, lis[1].find('a').text
    else:
        price, author = '', ''
    book_data.append((title, price, author))

# 1) CSV
with open('book.csv', 'w', newline='', encoding='utf-8-sig') as f:
    w = csv.writer(f); w.writerow(['title','price','author']); w.writerows(book_data)
# 2) JSON
with open('book.json', 'w', encoding='utf-8') as f:
    json.dump(book_data, f, ensure_ascii=False, indent=4)
print('CSV/JSON 저장 완료:', len(book_data), '권')
```
